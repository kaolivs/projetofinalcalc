```python
import numpy as np
import matplotlib.pyplot as plt

def verificar_dados(tempo, posicao):
    """
    Verifica se os dados de entrada so vlidos.

    Parmetros:
    tempo (list): Lista de tempos em segundos.
    posicao (list): Lista de posies em metros.

    Retorna:
    str: Mensagem de erro se algum problema for encontrado, caso contrrio None.
    """
    # Verifica se as listas tm o mesmo tamanho
    if len(tempo) != len(posicao):
        return "Erro: As listas de tempo e posio devem ter o mesmo tamanho."

    # Verifica se a lista de tempo est em ordem crescente
    if not all(tempo[i] <= tempo[i+1] for i in range(len(tempo)-1)):
        return "Erro: Os valores de tempo devem estar em ordem crescente."

    # Verifica se as listas no esto vazias
    if len(tempo) == 0 or len(posicao) == 0:
        return "Erro: As listas de tempo e posio no devem estar vazias."

    # Verifica se todos os elementos so numricos
    try:
        np.array(tempo, dtype=float)
        np.array(posicao, dtype=float)
    except ValueError:
        return "Erro: As listas devem conter apenas valores numricos."

    return None

def plotar_dados(tempo, posicao):
    """
    Plota os grficos de posio, velocidade e acelerao em funo do tempo.

    Parmetros:
    tempo (list): Lista de tempos em segundos.
    posicao (list): Lista de posies em metros.

    Retorna:
    str: Mensagem de erro se algum problema for encontrado, ou sucesso se os grficos forem gerados.
    """
    # Verificao dos dados de entrada
    erro = verificar_dados(tempo, posicao)
    if erro:
        return erro

    # Converso para arrays NumPy para facilitar os clculos
    tempo = np.array(tempo)
    posicao = np.array(posicao)

    # Clculo da velocidade (primeira derivada da posio)
    velocidade = np.gradient(posicao, tempo)

    # Clculo da acelerao (segunda derivada da posio)
    aceleracao = np.gradient(velocidade, tempo)

    # Criao da figura e dos subgrficos
    plt.figure(figsize=(12, 6))

    # Subgrfico 1: Posio vs Tempo
    plt.subplot(3, 1, 1)
    plt.plot(tempo, posicao, marker='o')
    plt.title('Posio vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posio (m)')

    # Subgrfico 2: Velocidade vs Tempo
    plt.subplot(3, 1, 2)
    plt.plot(tempo, velocidade, marker='o')
    plt.title('Velocidade vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Velocidade (m/s)')

    # Subgrfico 3: Acelerao vs Tempo
    plt.subplot(3, 1, 3)
    plt.plot(tempo, aceleracao, marker='o')
    plt.title('Acelerao vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Acelerao (m/s)')

    # Ajuste do layout para evitar sobreposio dos grficos
    plt.tight_layout()

    # Exibio dos grficos
    plt.show()
    return "Grficos gerados com sucesso."

# Exemplo de uso
# Listas de dados de tempo (em segundos) e posio (em metros)
tempo = [0, 1, 2, 3, 4, 5]
posicao = [0, 1, 4, 9, 16, 25]

# Chamada da funo para plotar os dados e exibir o resultado
resultado = plotar_dados(tempo, posicao)
if resultado:
    print(resultado)
```
