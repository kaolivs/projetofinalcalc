```python
import numpy as np
import matplotlib.pyplot as plt

def verificar_dados(tempo, posicao):
    """
    Verifica se os dados de entrada são válidos.

    Parâmetros:
    tempo (list): Lista de tempos em segundos.
    posicao (list): Lista de posições em metros.

    Retorna:
    str: Mensagem de erro se algum problema for encontrado, caso contrário None.
    """
    # Verifica se as listas têm o mesmo tamanho
    if len(tempo) != len(posicao):
        return "Erro: As listas de tempo e posição devem ter o mesmo tamanho."

    # Verifica se a lista de tempo está em ordem crescente
    if not all(tempo[i] <= tempo[i+1] for i in range(len(tempo)-1)):
        return "Erro: Os valores de tempo devem estar em ordem crescente."

    # Verifica se as listas não estão vazias
    if len(tempo) == 0 or len(posicao) == 0:
        return "Erro: As listas de tempo e posição não devem estar vazias."

    # Verifica se todos os elementos são numéricos
    try:
        np.array(tempo, dtype=float)
        np.array(posicao, dtype=float)
    except ValueError:
        return "Erro: As listas devem conter apenas valores numéricos."

    return None

def plotar_dados(tempo, posicao):
    """
    Plota os gráficos de posição, velocidade e aceleração em função do tempo.

    Parâmetros:
    tempo (list): Lista de tempos em segundos.
    posicao (list): Lista de posições em metros.

    Retorna:
    str: Mensagem de erro se algum problema for encontrado, ou sucesso se os gráficos forem gerados.
    """
    # Verificação dos dados de entrada
    erro = verificar_dados(tempo, posicao)
    if erro:
        return erro

    # Conversão para arrays NumPy para facilitar os cálculos
    tempo = np.array(tempo)
    posicao = np.array(posicao)

    # Cálculo da velocidade (primeira derivada da posição)
    velocidade = np.gradient(posicao, tempo)

    # Cálculo da aceleração (segunda derivada da posição)
    aceleracao = np.gradient(velocidade, tempo)

    # Criação da figura e dos subgráficos
    plt.figure(figsize=(12, 6))

    # Subgráfico 1: Posição vs Tempo
    plt.subplot(3, 1, 1)
    plt.plot(tempo, posicao, marker='o')
    plt.title('Posição vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição (m)')

    # Subgráfico 2: Velocidade vs Tempo
    plt.subplot(3, 1, 2)
    plt.plot(tempo, velocidade, marker='o')
    plt.title('Velocidade vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Velocidade (m/s)')

    # Subgráfico 3: Aceleração vs Tempo
    plt.subplot(3, 1, 3)
    plt.plot(tempo, aceleracao, marker='o')
    plt.title('Aceleração vs Tempo')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Aceleração (m/s²)')

    # Ajuste do layout para evitar sobreposição dos gráficos
    plt.tight_layout()

    # Exibição dos gráficos
    plt.show()
    return "Gráficos gerados com sucesso."

# Exemplo de uso
# Listas de dados de tempo (em segundos) e posição (em metros)
tempo = [0, 1, 2, 3, 4, 5]
posicao = [0, 1, 4, 9, 16, 25]

# Chamada da função para plotar os dados e exibir o resultado
resultado = plotar_dados(tempo, posicao)
if resultado:
    print(resultado)
```
